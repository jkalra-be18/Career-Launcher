Q9. Plot the 21 day and 34 day Moving average with the average price and decide a Call ! Call should be buy whenever the smaller moving average (21) crosses over longer moving average (34) AND the call should be sell whenever smaller moving average crosses under longer moving average. One of the most widely used technical indicators.

```python
df9 = pd.DataFrame()
df9["SMA_21"] = data["Close Price"].rolling(21).mean()
df9["LMA_34"] = data["Close Price"].rolling(34).mean()

column_names = df9.columns
plt.figure(figsize=(20,8))
plt.plot(df9)
plt.legend(column_names)
plt.show()
```

```python
df9["Close Price"] = data["Close Price"]
df9.tail()
```

```python
p_index       = df9.index[0]
p_row         = df9[:p_index]
p_sma = p_row["SMA_21"][0] > p_row["LMA_34"][0]
df9['Trade_Call'] = 'HOLD'

for index, row in df9.iterrows():
    if row["SMA_21"] and row["LMA_34"]:    
        if p_sma and row["SMA_21"] < row["LMA_34"]:
            result = "SELL"
        elif not p_sma and row["SMA_21"] > row["LMA_34"]:
            result = "BUY"
        else:
            result = "HOLD" 
        df9.at[index, 'Trade_Call'] = result
        p_sma = row["SMA_21"] > row["LMA_34"]                        
    
df9_Trade_Call = df9[df9['Trade_Call'] != "HOLD" ]
df9_Trade_Call
```

```python
fig = plt.figure(figsize=(20,10))
ax1 = fig.add_subplot(111, ylabel="Price")

df9[["Close Price", "SMA_21", "LMA_34"]].plot(ax=ax1)

ax1.plot( df9["Close Price"][ df9['Trade_Call'] == 'BUY' ].index, 
          df9["Close Price"][ df9['Trade_Call'] == 'BUY' ], 
          '^', markersize=15, color='g' )

ax1.plot( df9["Close Price"][ df9['Trade_Call'] == 'SELL' ].index, 
          df9["Close Price"][ df9['Trade_Call'] == 'SELL' ], 
          'v', markersize=15, color='r' )
```

Q10. Trade Calls - Using Bollinger Bands
Plot the bollinger bands for this stock - the duration of 14 days and 2 standard deviations away from the average The bollinger bands comprise the following data points-

The 14 day rolling mean of the closing price (we call it the average)
Upper band which is the rolling mean + 2 standard deviations away from the average.

Lower band which is the rolling mean - 2 standard deviations away from the average.

Average Daily stock price.

Bollinger bands are extremely reliable , with a 95% accuracy at 2 standard deviations , and especially useful in sideways moving market. Observe the bands yourself , and analyse the accuracy of all the trade signals provided by the bollinger bands. Save to a new csv file.

```python
data["14_day_SMA"] = data["Close Price"].rolling(14).mean()
data["14_day_STD"] = data["Close Price"].rolling(14).std()
data["14_day_bollinger_upper"] = data["14_day_SMA"] + data["14_day_STD"] * 2
data["14_day_bollinger_lower"] = data["14_day_SMA"] - data["14_day_STD"] * 2
```

```python
plt.figure(figsize=(25,15))
ax = plt.gca()

data.plot(kind='line',x='Date', y='Average Price', ax=ax, color='black')
data.plot(kind='line',x='Date', y='14_day_SMA',    ax=ax, color='blue', linestyle='--')
data.plot(kind='line',x='Date', y='14_day_bollinger_upper',    ax=ax, color='red', linestyle='--')
data.plot(kind='line',x='Date', y='14_day_bollinger_lower',    ax=ax, color='g', linestyle='--')

plt.xlabel('Date', size = 15)
plt.title("14 day bollinger bands", size = 15)
plt.ylabel("Price", size = 15)
plt.show()
```
