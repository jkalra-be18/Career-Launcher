# 1.4 In a separate array , calculate the monthwise VWAP (Volume Weighted Average Price ) of the stock. ( VWAP = sum(price*volume)/sum(volume) ). The values for this column can be derived from the ‘Date” column by using appropriate pandas functions. Similarly, create a column ‘Year’ and initialize it. Then use the 'groupby()' function by month and year. Finally, calculate the vwap value for each month (i.e. for each group created).

```python
df1 = pd.concat([df1, pd.DataFrame({'Numerator': df1['Average Price']*df1['Total Traded Quantity']})], axis = 1)
```

```python
month_wise_VWAP = df1['Numerator'].groupby(df1['Date'].
                                           dt.strftime('%B')).sum() / df1['Total Traded Quantity'].groupby(df1['Date'].dt.strftime('%B')).sum()
print(month_wise_VWAP)
```

```python
year_wise_VWAP = df1['Transactions'].groupby(df1['Date'].
                                                      dt.strftime('%Y')).sum() / df1['Total Traded Quantity'].groupby(df1['Date'].dt.strftime('%Y')).sum()
print(year_wise_VWAP)
```

# 1.5 Write a function to calculate the average price over the last N days of the stock price data where N is a user defined parameter. Write a second function to calculate the profit/loss percentage over the last N days. Calculate the average price AND the profit/loss percentages over the course of last - 1 week, 2 weeks, 1 month, 3 months, 6 months and 1 year. {Note : Profit/Loss percentage between N days is the percentage change between the closing prices of the 2 days }

```python
def average_price(n, stmp = 'D'):
    
    return df1['Close Price'].last(str(n) + stmp).mean()
average_price(90)
```

```python
def profit_loss(n, stmp = 'D'):
    k = len(df1)
    curr = df1['Close Price'][k-1]
    n_days_old = df1['Close Price'].last(str(n) + stmp)[0]
    
    return ((curr - n_days_old)/n_days_old)*100
profit_loss(90)
```

```python
def average_price_and_profit_loss(num, stmp):
    
    print("Average Price over last " + str(num) + stmp+ " is", average_price(num, stmp))

    val = profit_loss(num, stmp)

    if val == 0.0:
        print("No profit and No loss")
    elif val > 0:
        print("Profit over last {} {} is {:.2f}%".format(num, stmp, val))
    else:
        print("Loss over last {} {} is {:.2f}%".format(num, stmp, val*-1))
        
# Average price and profit/loss over last 1 week
print('Over last Week:',average_price_and_profit_loss(1, 'W'))


# Average price and profit/loss over last 2 week
print('Over last 2 Weeks:',average_price_and_profit_loss(2, 'W'))


# Average price and profit/loss over last 1 month
print('Over last Month:',average_price_and_profit_loss(1, 'M'))


# Average price and profit/loss over last 3 month
print('Over last 3 Months',average_price_and_profit_loss(3, 'M'))


# Average price and profit/loss over last 6 month
print('Over last 6 Months:',average_price_and_profit_loss(6, 'M'))


# Average price and profit/loss over last 1 year
print('Over last 1 Year:',average_price_and_profit_loss(1, 'Y'))
```

# 1.6 Add a column 'Day_Perc_Change' where the values are the daily change in percentages i.e. the percentage change between 2 consecutive day's closing prices. Instead of using the basic mathematical formula for computing the same, use 'pct_change()' function provided by Pandas for dataframes. # 1.7 Add another column 'Trend' whose values are: 'Slight or No change' for 'Day_Perc_Change' in between -0.5 and 0.5 'Slight positive' for 'Day_Perc_Change' in between 0.5 and 1 'Slight negative' for 'Day_Perc_Change' in between -0.5 and -1 'Positive' for 'Day_Perc_Change' in between 1 and 3 'Negative' for 'Day_Perc_Change' in between -1 and -3 'Among top gainers' for 'Day_Perc_Change' in between 3 and 7 'Among top losers' for 'Day_Perc_Change' in between -3 and -7 'Bull run' for 'Day_Perc_Change' >7 'Bear drop' for 'Day_Perc_Change' <-7

```python
df1.insert(df1.shape[1], 'Trend', 'NaN')
df1.head()
```

```python
# using .loc to update the value in the column
df1.loc[df1['Day_Perc_Change'].between(-0.005, -0.01, inclusive = True), 'Trend'] = 'Slight negative'

df1.loc[df1['Day_Perc_Change'].between(0.005, 0.01, inclusive = True), 'Trend'] = 'Slight positive'
df1.loc[df1['Day_Perc_Change'].between(-0.03, -0.01, inclusive = False),'Trend'] = 'Negative'

df1.loc[df1['Day_Perc_Change'].between(0.01, 0.03, inclusive = False),'Trend'] = 'Positive'
df1.loc[df1['Day_Perc_Change'].between(-0.07, -0.03, inclusive = True),'Trend'] = 'Top losers'

df1.loc[df1['Day_Perc_Change'].between(0.03, 0.07, inclusive = True),'Trend'] = 'Top gainers'
df1.loc[df1['Day_Perc_Change']<-0.07,'Trend']= 'Bear drop'

df1.loc[df1['Day_Perc_Change']>0.07,'Trend'] = 'Bear run'
df1.head()
```
